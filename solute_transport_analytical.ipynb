## Solute Transport

```python
#PSP_soluteTransportAnalytical
import numpy as np
from scipy import special as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tmax = 11    
nrHours = 10                                       
n = 100                                                         

#dispersion coefficient
D = 1. 
D1 = 0.25

#water flux density
fw = 2.
fw1 = 1.

#velocity
v = 0.5
v1 = 1.5

mass = 5.0
z = np.zeros(n)
conc = np.zeros(n) 
conc1 = np.zeros(n) 
sqrtPi = np.sqrt(np.pi) 
for i in range(n):
    z[i] = i
```

```python
# create a figure and axes
fig = plt.figure(figsize=(5,5))
ax1 = plt.subplot(1,1,1)   

plt.ion()
conc[0] = 0.0
conc1[0] = 0.
t=1
for i in range(1, n):
    a = v / (sqrtPi*np.sqrt(D*t)) * np.exp(-((z[i] - v*t)**2) / (4.0*D*t))
    b = (v**2)/(2*D) * np.exp((v*z[i])/D) * sp.erfc((z[i]+v*t) / np.sqrt(4.0*D*t))
    conc[i]= mass/fw * (a - b)                  
    
    a1 = v1 / (sqrtPi*np.sqrt(D1*t)) * np.exp(-((z[i] - v1*t)**2) / (4.0*D1*t))
    b1 = (v1**2)/(2*D1) * np.exp((v1*z[i])/D1) * sp.erfc((z[i]+v1*t) / np.sqrt(4.0*D1*t))
    conc1[i]= mass/fw1 * (a1 - b1)                  


ax1.set_xlabel('Concentration [g cm$^{-2}$]',fontsize=16,labelpad=8)
ax1.set_ylabel('Depth [cm]',fontsize=16,labelpad=8)
#ax1.set_xlim(0, 0.4)
ax1.set_ylim(-50, 2)
ax1.plot([0,0.4],[0,0],'k:',lw=1)
line1, = ax1.plot(conc, -z, 'k-')
line2, = ax1.plot(conc1, -z, '-', c='steelblue')
```

```python
conc[0] = 0.0
def drawframe(t):
    for i in range(1, n):
        a = v / (sqrtPi*np.sqrt(D*t)) * np.exp(-((z[i] - v*t)**2) / (4.0*D*t))
        b = (v**2)/(2*D) * np.exp((v*z[i])/D) * sp.erfc((z[i]+v*t) / np.sqrt(4.0*D*t))
        conc[i]= mass/fw * (a - b) 
        
        a1 = v1 / (sqrtPi*np.sqrt(D1*t)) * np.exp(-((z[i] - v1*t)**2) / (4.0*D1*t))
        b1 = (v1**2)/(2*D1) * np.exp((v1*z[i])/D1) * sp.erfc((z[i]+v1*t) / np.sqrt(4.0*D1*t))
        conc1[i]= mass/fw1 * (a1 - b1)                  
        
    line1.set_data(conc, -z)
    line2.set_data(conc1, -z)
    return (line1,line2)
```

```python
from matplotlib import animation

# blit=True re-draws only the parts that have changed.
anim = animation.FuncAnimation(fig, drawframe, frames=30, interval=300, blit=True)
```
